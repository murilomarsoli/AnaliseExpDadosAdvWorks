# Análise Exploratória da Base de Dados "Adventure Works"

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
df = pd.read_excel("/content/AdventureWorks.xlsx")
```

```python
df.head()
```

```python
# Quantidade de linhas e colunas
df.shape
```

```python
# Verificar o tipo de dados de cada coluna
df.dtypes
```

```python
# Verificar se temos valores nulos no df

df.isnull().sum()
```

```python
# Qual é a receita total da empresa?

df["Valor Venda"].sum()
```

```python
# Qual é o custo total de cada venda que a empresa tem com a fabricação de seus produtos?

df["Custo Venda"] = df["Custo Unitário"] * df["Quantidade"]
```

```python
df.head(2)
```

```python
# Qual é o custo total da empresa?

df["Custo Venda"].sum()
```

```python
df.head(2)
```

```python
# Tendo em vista a receita e custos gerados em cada venda, qual é o lucro que cada venda gera?

df["Lucro Venda"] = df["Valor Venda"] - df["Custo Venda"]
```

```python
df.head(2)
```

```python
# Qual é o lucro total da empresa?

df["Lucro Venda"].sum()
```

```python
round(df["Lucro Venda"].sum(), 2) # round(xxx, 2) - arredondar em duas casas decimais
```

```python
# Criando uma coluna para armazenar o total de dias para o envio do pedido

df["Tempo Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(2)
```

```python
# Extrair somente os dias (número de dias) do tempo de envio (retirar o "days" do final)

df["Tempo Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(2)
```

```python
# Renomear a coluna "Tempo Envio" para "Tempo Dias Envio"

df = df.rename(columns={"Tempo Envio":"Tempo Dias Envio"})
```

```python
df.head(2)
```

```python
# Verificar o tipo da coluna "Tempo Dias Envio"

df["Tempo Dias Envio"].dtype
```

```python
# Média de dias para envio dos pedidos por marca

df.groupby("Marca")["Tempo Dias Envio"].mean()
```

```python
# Lucro agrupado por ano e por marca

df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Venda"].sum()
```

```python
# Configurar display de números float para mostrar os números de notação científica em números decimais

pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Lucro agrupado por ano e por marca (com a config acima aplicada)

df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Venda"].sum()
```

```python
# Resetando o index
# Lucro agrupado por ano e por marca de forma tabular

lucro_ano_marca = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Venda"].sum().reset_index()
lucro_ano_marca
```

```python
# Qual é a quantidade total de produtos vendidos pela empresa?

df["Quantidade"].sum()
```

```python
# Qual é a quantidade de produtos vendidos por cada produto?

df.groupby("Produto")["Quantidade"].sum()
```

```python
# Qual é a quantidade de produtos vendidos por cada produto?
# Ordenados pelo produto que mais vendeu ao que menos vendeu

df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
# Gráfico da quantidade de produtos vendidos por cada produto
# Ordenado pelo produto que mais vendeu ao que menos vendeu

df.groupby("Produto")["Quantidade"].sum().sort_values().plot.barh(title="Quantidade de unidades vendidas x produto")
plt.xlabel("Quantidade")
plt.ylabel("Produto")
```

```python
# Gráfico do lucro gerado pela empresa em cada ano
# Ordenado pelo ano e de forma crescente

df.groupby(df["Data Venda"].dt.year)["Lucro Venda"].sum().plot.bar(title="Lucro x ano")
plt.xlabel("Ano")
plt.ylabel("Lucro")
plt.legend()
# os valores são mostrados em notação científica
```

```python
# Lucro gerado pela empresa em cada ano (para observarmos os valores em forma decimal)
# Ordenado pelo ano e de forma crescente

df.groupby(df["Data Venda"].dt.year)["Lucro Venda"].sum()
```

```python
# Filtrar somente as vendas que foram feitas no ano de 2009

df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
# Qual foi o lucro mensal da empresa no ano de 2009?

df_2009.groupby(df["Data Venda"].dt.month)["Lucro Venda"].sum()
```

```python
# Gráfico de linha do lucro mensal da empresa no ano de 2009

df_2009.groupby(df["Data Venda"].dt.month)["Lucro Venda"].sum().plot(title="Lucro x Mês (2009)", marker="o")
plt.xlabel("Mês")
plt.ylabel("Lucro")
plt.legend()
```

```python
df.dtypes
```

```python
# Lucro que cada marca gerou para a empresa no ano de 2009

df_2009.groupby("Marca")["Lucro Venda"].sum()
```

```python
# Gráfico para visualização do lucro que cada marca gerou para a empresa no ano de 2009

df_2009.groupby("Marca")["Lucro Venda"].sum().plot.bar(title="Lucro x Marca (2009)")
plt.xlabel("Marca")
plt.ylabel("Lucro")

# Legendas do eixo x (vertical por padrão)
plt.xticks(rotation="horizontal")
```

```python
# Lucro que cada classe está gerando a empresa

df.groupby("Classe")["Lucro Venda"].sum()
```

```python
# Gráfico para visualização do lucro que cada classe está gerando à empresa

df.groupby("Classe")["Lucro Venda"].sum().plot.bar(title="Lucro x Classe")
plt.xticks(rotation="horizontal")
plt.xlabel("Classe")
plt.ylabel("Lucro (em notação científica)")
```

```python
# Análise estatística do tempo (dias) para envio dos pedidos

df["Tempo Dias Envio"].describe()

# count - 904 registros na coluna
# mean - média
# std - desvio padrão
# min - valor mínimo
# 25 % - 25% dos dados (1/3 dos dados tem o tempo de envio de até 6 dias) - 1º QUARTIL
# 50% - 50% dos dados (1/2 dos dados tem o tempo de envio de até 9 dias) - 2º QUARTIL
# 75% - 75% dos dados (3/4 dos dados tem o tempo de envio de até 11 dias) - 3º QUARTIL
# max - valor máximo
```

```python
# Gráfico de Boxplot ("Análise Estatística")

plt.boxplot(df["Tempo Dias Envio"]);

# Valor "Outlier" -> valor máximo mas que é muito discrepante, fora do padrão (bolinha preta lá em cima) = 20
# Limite superior -> valor máximo não-outlier/dentro do padrão (linhazinha preta superior) = 13
# Q3 -> 3º QUARTIL (última linha preta da caixa) = 11
# Q2 -> 2º QUARTIL e Mediana (linha verde do meio da caixa) = 9
# Q1 -> 1º QUARTIL (primeira linha preta da caixa) = 6
# Limite inferior -> valor mínimo (linhazinha preta inferior) = 4
```

```python
# Histograma do tempo (dias) para envio dos pedidos

plt.hist(df["Tempo Dias Envio"]);
plt.xlabel("Dias para entrega")
plt.ylabel("Vendas")
```
